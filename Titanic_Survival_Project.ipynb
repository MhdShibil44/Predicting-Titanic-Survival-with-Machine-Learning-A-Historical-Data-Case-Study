```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Loading the dataset
df = pd.read_csv(r'D:\ExcelR\Project\DataScience_Project\Titanic survivors logistic regression\Titanic-Dataset.csv')
df
```

```python
# To drop unwanted columns and save the data as the original dataset. 
df = df.drop(['PassengerId','Name','Ticket'],axis=1)
df
```

```python
# To check the type of data and missing values
df.info()
```

```python
# To get the total null values
df.isnull().sum()
```

```python
# To get the percnetage of null values
df.isnull().sum()/len(df)*100
```

```python
# Dropping the Cabin column as most of its values are null.
df = df.drop(['Cabin'],axis = 1)
```

```python
# To fill the null values
# Here we take mean first to fill the null values in Age, but since it did not perform well in training and testing so we use the median instead.

df['Age'] = df['Age'].fillna(round(df['Age'].median()))
```

```python
# To get the count of each element in the Embarked data.
df['Embarked'].value_counts()
```

```python
# To fill the null values in the Embarked data with the mode element.
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
# To check the percentage of remaining null values after filling the Embarked data with the mode element.
df.isnull().sum()/len(df)*100
```

```python
# To convert the textual data in the Sex and Embarked columns using feature encoding (here we use One Hot Encoding).
df = pd.get_dummies(data=df,columns=['Sex','Embarked'],dtype=int,drop_first=True)
df
```

```python
# Assigning x_data and y_data
x = df.drop(['Survived'],axis=1)
y = df['Survived']
```

```python
# To spit the data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Assigning logisticRegression to a variable
lr = LogisticRegression()
lr
```

```python
# To fit the data in LR
lr.fit(x_train,y_train)
```

```python
# To get the m value or coefficient
lr.coef_
```

```python
# To get the intercept
lr.intercept_
```

```python
# To get the predicted y_train and y_test
y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)
```

```python
# To get the classification report including accuracy, precision, recall, and F1-score for the training data.
print(classification_report(y_train,y_pred_train))
```

```python
# To get the classification report including accuracy, precision, recall, and F1-score for the training data.
print(classification_report(y_test,y_pred_test))
```
